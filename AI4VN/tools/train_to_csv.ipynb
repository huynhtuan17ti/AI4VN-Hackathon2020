```python
import cv2
import matplotlib.pyplot as plt
import os
import imutils 
import skimage
import skimage.transform
import pandas as pd
```

```python
IMAGE_PATH = "/home/ryan/Machine_Learning/AI4VN/Images"
SAVE_PATH = "/home/ryan/Machine_Learning/AI4VN/train"
IMG_SIZE = 800
NUM_CLASSES = 8
```

```python
for i in range(8):
    class_path = os.path.join(IMAGE_PATH, "Label_{}".format(i))
    folder_image = os.listdir(class_path)
    cnt = 0
    for x in folder_image:
        image_path = os.path.join(class_path, x)
        lst = os.listdir(image_path)
        cnt += len(lst)
    print('Number image of class "{}" : {}'.format(i, cnt))
```

```python
def resize_keep_aspect(img, shape):
    h, w, _ = img.shape
    maxHW = max(h,w)
    top = (maxHW-h)//2
    bottom = (maxHW-h) - top
    left = (maxHW-w)//2
    right = (maxHW-w) - left
    withBorder = cv2.copyMakeBorder(img, top, bottom, left, right, cv2.BORDER_CONSTANT)
    return skimage.transform.resize(withBorder, shape)
```

```python
image_id = []
label = []
image_cnt = 0
for i in range(8):
    class_path = os.path.join(IMAGE_PATH, "Label_{}".format(i))
    folder_image = os.listdir(class_path)
    for x in folder_image:
        image_path = os.path.join(class_path, x)
        lst = os.listdir(image_path)
        for img in lst:
            path = os.path.join(image_path, img)
            try:
                type_img = path[-4:]
                if type_img == 'jpeg':
                    type_img = '.jpeg'
                if type_img == '.bmp':
                    print(img)
                    continue
                img_array = cv2.imread(path)
                h, w, _ = img_array.shape
                #img_array = resize_keep_aspect(img_array, (IMG_SIZE, IMG_SIZE))
                
                #print(image_id)
                cv2.imwrite(os.path.join(SAVE_PATH , str(image_cnt) + type_img), img_array)
                image_id.append(str(image_cnt) + type_img)
                label.append(i)
                image_cnt += 1
            except Exception as E:
                print(E)
                print('Error {} {}'.format(i, img))
                pass
```
